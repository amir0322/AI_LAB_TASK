```python
X = df.drop(columns=['Water_Consumed_Liters'])
y = df['Water_Consumed_Liters']

X['Date'] = pd.to_datetime(X['Date'])
X['Day'] = X['Date'].dt.day
X['Month'] = X['Date'].dt.month
X['Year'] = X['Date'].dt.year
X = X.drop(columns=['Date'])

X['City'] = X['City'].str.lower()

cat_columns = X.select_dtypes(['object']).columns
X[cat_columns] = X[cat_columns].apply(lambda col: pd.factorize(col)[0])
```

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X,y, test_size=0.3, shuffle=False)
```

```python
print(train_X.shape, train_y.shape)
print(test_X.shape, test_y.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor # Support Vector Classifier

model_rfr = RandomForestRegressor()
model_rfr.fit(train_X, train_y)

print(model_rfr)
```

```python
model_pred_rfr = model_rfr.predict(test_X)
model_pred_rfr
```

```python
from sklearn.metrics import r2_score
y_pred = model_rfr.predict(test_X)
model_r2 = r2_score(test_y, y_pred)
print("Here is model accuracy score (R²):", round(model_r2, 3)*100, "%")
```
